```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/leaderboards.csv', index_col=1)
df.index = pd.to_datetime(df.index)
```

```python
df
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 7))
ax[0].hist(df['score'], bins=20, density=True)
ax[1].plot(df['score'])
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 7))
sns.countplot(data=df, x='tier', ax=ax[0])
ax[1].plot(df['tier'])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
h = ax.hist2d(df['tier'], df['score'], bins=[len(df['tier'].unique()), 20])
fig.colorbar(h[3], ax=ax)
ax.set_ylabel('Score')
ax.set_xlabel('Tier')
```
